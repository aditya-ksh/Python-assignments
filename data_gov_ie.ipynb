```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
```

We have already downloaded the CSV data file from https://data.gov.ie/dataset/homelessness-report-december-2023.

This data files homelessness figures for different regions and in Ireland.

Using `pandas`, we load this data into a `DataFrame`

```python
data = pd.read_csv('homelessness-report-december-2023.csv', index_col="Region")
data
```

We find the number of rows and columns using the `shape` attribute of the `DataFrame`. There are 9 rows and 18 columns. Notice that we have used the first column, named `Region`, in our original dataset as the `Index`.

```python
data.shape
```

We perform some exploratory data analysis using to the `describe()` function to find out summary statistics.

```python
data.describe()
```

We can plot the data from a column using the `matplotlib` package. In this case, first we extract the __Total Adults__ column. Then, we are using the `barh` function to create a horizontal bar chart where the labels for the y-axis are provided by the `Index` and x-axis by the `ndarray` of values in the extracted column.

```python
# Extract on columns as a Series
tAdults = data["Total Adults"]

# Create a horizontal bar chart using index and values
plt.barh(tAdults.index, tAdults.values)

# Display the plot
plt.show()
```
